## 2、K 最近邻算法
K 最近邻算法（K-Nearest Neighbors，KNN）
可用于分类和回归，选择离数据点最近的若干个训练数据，计算距离这些点的距离平均值作为预测值

```python
from sklearn.datasets import make_blobs
# 导入 KNN 分类器
from sklearn.neighbors import KNeighborsClassifier

# 导入数据集拆分工具
from sklearn.model_selection import train_test_split

# 生成样本数为 200，分类数为 2 的训练数据集
data1 = make_blobs(n_samples=200, centers = 2,random_state=8)
x,y=data1
# 将生成的数据进行可视化
plt.scatter(x[:,0],x[:,1],c=y,cmap=plt.cm.spring,edgecolor='k')
```

```python
clf = KNeighborsClassifier()
#拟合模型
clf.fit(x,y)
```

```python
x_min,x_max=x[:,0].min()-1 ,x[:,0].max()+1
y_min,y_max=x[:,1].min()-1 ,x[:,1].max()+1

xx, yy = np.meshgrid(np.arange(x_min,x_max, .02),
                     np.arange(y_min,y_max, .02))

#使用模型进行预测
z= clf.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)

#绘制模型预测区域
plt.pcolormesh(xx, yy, z,cmap=plt.cm.Pastel1)
# 绘制训练数据
plt.scatter(x[:,0], x[:,1],c=y, cmap=plt.cm.spring, edgecolor='k')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

# 标记一个需要预测的数据
plt.scatter(6.75, 4.82, marker='*', c='red',s=200)

plt.title("Classifier:KNN")
```

### K最近邻算法用于回归分析

```python
# 导入回归分析数据集生成器
from sklearn.datasets import make_regression

# 生成特征数量为1，噪音为 50 的数据集
X, y = make_regression(n_features=1, n_informative=1, noise=50, random_state=8)

plt.scatter(X,y,c='orange',edgecolor='k')
plt.show()
```

## 线性模型的一般公式

线性模型通常写作：



$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_p x_p + \varepsilon
$$



其中：
- $y$：因变量（响应变量、目标变量）
- $\beta_0$：截距项
- $x_i$：第 $i$ 个自变量（特征）
- $\beta_i$：第 $i$ 个自变量的系数（回归系数，weights）
- $\varepsilon$：误差项


```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# 用线性模型拟合，两个点（1,3），（4,5）
X=[[1],[4]]
y=[3,5]

# 用线性回归模型拟合
lr = LinearRegression().fit(X,y)

z=np.linspace(0,5,20)
plt.scatter(X,y,s=80)

plt.plot(z, lr.predict(z.reshape(-1,1)),c='k')
plt.title('straight line')
plt.show()
```

```python
# 获取模型的截距和系数
print(lr.intercept_, lr.coef_[0])

# 模型就是线性方程：y=2.33 + 0.667x
```

```python
# 线性模型拟合拟合三个点
X=[[1],[4],[3]]
y=[3,5,3]
lr=LinearRegression().fit(X,y)
z=np.linspace(0,5,20)

plt.scatter(X,y,s=80)
plt.plot(z, lr.predict(z.reshape(-1,1)),c='k')
plt.title('Straight Line & 3 points')
plt.show()
```

```python
# 拟合的结果：拟合出的线性模型为距离每个数据点的加和最小的直线（线性回归模型的原理）
# 获取模型的截距和系数
print(lr.intercept_, lr.coef_[0])

# 模型就是线性方程：y=2.143 + 0.571x
```

```python
from sklearn.datasets import make_regression
X,y=make_regression(n_samples=50,n_features=1,n_informative=1,
                    noise=50,random_state=1)

reg=LinearRegression()
reg.fit(X,y)

z=np.linspace(-3,3,200).reshape(-1,1)
plt.scatter(X,y,c='b',s=60)
plt.plot(z,reg.predict(z),c='k')
plt.title('Linear Regression')
```

```python
# 拟合的结果：拟合出的线性模型为距离每个数据点的加和最小的直线（线性回归模型的原理）
# 获取模型的截距和系数
print(reg.intercept_, reg.coef_[0])

# 模型就是线性方程：y=10.92 + 79.52x
```

- 对于特征数为 1 的数据集，线性回归模型是直线；
- 对于特征数为 2 的数据集，线性回归模型是平面；
- 对于更多特征数量的数据集，线性回归模型是高纬度的超平面。

使用线性模型的前提条件是假设目标 y 是数据特征的线性组合；
尽管使用一维数据作为练习时模型没有覆盖很多的特征数据点，对于特征变量较多的数据集线性模型十分强大；
尤其是，训练数据集的特征变量大于数据点的数量时，线性模型可以对训练数据做出近乎完美的预测。

用于回归的线性模型有很多种，其区别在于如何从训练数据确定回归系数 $\beta$ 和截距 $\beta_0$


### 最基本的线性模型——线性回归
线性回归，也称为普通最小二乘法（OLS），原理是找到训练数据集的预测值和其真实值的平方差最小时所对应的回归系数 $\beta$ 和截距 $\beta_0$。

线性回归**没有可调的参数**——简单但是**无法控制模型的复杂性**。

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X,y=make_regression(n_samples=100,n_features=2,n_informative=2,random_state=38)
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=8)
lr = LinearRegression().fit(X_train,y_train)


# 获取模型的截距和系数（有 2 个特征，系数的个数也是 2）
print(lr.intercept_, lr.coef_[:])

#模型得分
print(lr.score(X_train,y_train))
print(lr.score(X_test,y_test))
```

上例由于没有往数据集添加 noise，得分是满分；

使用真实世界数据集（糖尿病数据集）再测试一次

```python
from sklearn.datasets import load_diabetes

X,y = load_diabetes().data,load_diabetes().target
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=8)

lr = LinearRegression().fit(X_train,y_train)
#模型得分
print(lr.score(X_train,y_train))
print(lr.score(X_test,y_test))
```

线性回归模型很容易出现**过拟合**的现象——训练得分很高，测试得分很低

### 使用 L2 正则化的线性模型——岭回归
岭回归实际是一种改良的最小二乘法，是一种能够避免过拟合的线性模型。

在岭回归中，模型会保留所有的特征变量，但是会减小特征变量的系数值——让特征变量对预测结果的影响变小。

岭回归通过改变其 alpha 参数来控制减小特征系数的程度。

这种保留所有特征变量，同时减小特征变量系数(不为 0)来避免过拟合的方法就是**L2 正则化**

```python
# 基于相同的糖尿病数据集进行岭回归模型的训练
from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train,y_train)

#模型得分
print(ridge.score(X_train,y_train))
print(ridge.score(X_test,y_test))
```

虽然岭回归模型的训练数据集得分比线性回归模型的训练数据集得分低，但是其测试数据集得分和训练数据集得分非常相近。

岭回归是一个相对受限的模型，发生过拟合的可能性大大降低了；可以说，复杂度越低的模型，在训练数据集上的表现越差，但是其泛化能力更好。

如果更在意模型的泛化能力应该选择岭回归，而不是线性回归。

#### 岭回归模型参数调节

岭回归是在模型的简单性（使系数趋近于 0）和在训练数据集上的性能表现之间取得平衡的模型。

可以使用 alpha 参数（默认 1）控制使模型更简单还是使在训练数据集上性能表现更好。

增加 alpha 参数会减小特征变量系数，从而降低其在训练集的表现，但是有助于泛化。因此当模型出现过拟合时，可以增加 alpha 参数值。

```python
# 依然使用前例的数据集，更改 alpha 值
ridge10 = Ridge(alpha=10).fit(X_train,y_train)

#模型得分
print(ridge10.score(X_train,y_train))
print(ridge10.score(X_test,y_test))

print('-------------')
ridge01 = Ridge(alpha=0.1).fit(X_train,y_train)

#模型得分
print(ridge01.score(X_train,y_train))
print(ridge01.score(X_test,y_test))
```

通过上例可见，大的 alpha 参数值使模型泛化能力更强（甚至出现测试集得分高于训练集得分的情况）；

而当 alpha 参数值很小时，模型得分和线性回归模型非常接近。

```python
import matplotlib.pyplot as plt

# 绘制alpha=1时的模型系数
plt.plot(ridge.coef_, 's', label='Ridge alpha=1')
# 绘制alpha=10时的模型系数
plt.plot(ridge10.coef_, '^', label='Ridge alpha=10')
# 绘制alpha=0.1时的模型系数
plt.plot(ridge01.coef_, 'v', label='Ridge alpha=0.1')
# 绘制线性回归的系数作为对比
plt.plot(lr.coef_, 'o', label='linear regression')
# 系数的索引
plt.xlabel("coefficient index")
# 系数的数量级
plt.ylabel("coefficient magnitude")
plt.hlines(0,0,len(lr.coef_))
# 显示图例
plt.legend()
```

线性模型的系数是数组，上图中显示了不同模型各个系数的数量级。

可见，随着 alpha 参数的递增，系数量级趋近于 0。线性回归模型没有正则化处理，其特征变量的系数的数量级非常大

```python
import numpy as np
from sklearn.model_selection import learning_curve, KFold

def plot_learning_curve(est, X, y):
    # cv 交叉验证
    training_sizes, train_scores, test_scores = learning_curve(
        est, X, y, train_sizes=np.linspace(.1, 1, 20), cv=KFold(20, shuffle=True, random_state=1)
    )
    estimator_name = est.__class__.__name__
    line = plt.plot(training_sizes, train_scores.mean(axis=1), '--', label="training " + estimator_name)
    plt.plot(training_sizes, test_scores.mean(axis=1), '-', label="test " + estimator_name, c=line[0].get_color())
    plt.xlabel('Training set size')
    plt.ylabel('Score')
    plt.ylim(0, 1.1)

plot_learning_curve(Ridge(alpha=1), X, y)
plot_learning_curve(LinearRegression(), X, y)
plt.legend(loc=(0, 1.05), ncol=2, fontsize=11)
```

学习曲线（learning_curve）函数用于评估模型在不同训练集大小下的性能，返回训练集样本数、训练分数和验证分数。

岭回归是经过正则化的模型，得分始终比线性回归低。岭回归在测试集和训练集的得分差距较小。

随着数据集规模增加，模型表现都越来越好。线性回归模型的测试集和训练集的得分差距也逐渐缩小（不容易出现过拟合）。可见，如果数据规模足够大正则化就不那么重要了。

### 使用 L1 正则化的线性模型——套索回归

**套索（Lasso）** 回归使用 L1 正则化，也是将特征系数限制在 0 附近（与 L2 正则化相同），但是有部分特征系数恰好会是 0（与 L2 正则化不同）。即，套索回归会突出重要的特征，忽略不重要特征。

```python
from sklearn.linear_model import Lasso

# 同样使用前例的糖尿病数据集进行训练
lasso = Lasso().fit(X_train, y_train)

#模型得分
print(lasso.score(X_train,y_train))
print(lasso.score(X_test,y_test))
# 模型使用的参数个数（即 特征系数不为 0 的特征数）
print(np.sum(lasso.coef_!=0))
```

从上面的结果可见，模型只用了三个特征，得分很差——出现了**欠拟合**的问题。

#### 套索回归的参数调节
套索回归的参数：
- alpha：控制特征系数被约束为 0 的强度，默认值 1。降低 alpha 值会拟合出更复杂的模型，但是如果太低会倾向于出现过拟合。
- max_iter：最大迭代次数

```python
# alpha=0.1，得分表现更好
lasso01 = Lasso(alpha=0.1,max_iter=100000).fit(X_train, y_train)

#模型得分
print(lasso01.score(X_train,y_train))
print(lasso01.score(X_test,y_test))
# 模型使用的参数个数
print(np.sum(lasso01.coef_!=0))
```

```python
# alpha=0.0001，训练集得分和测试集得分差距变大——有过拟合倾向
lasso00001 = Lasso(alpha=0.0001,max_iter=100000).fit(X_train, y_train)

#模型得分
print(lasso00001.score(X_train,y_train))
print(lasso00001.score(X_test,y_test))
# 模型使用的参数个数
print(np.sum(lasso00001.coef_!=0))
```

```python
# 绘制alpha值等于1时的模型系数
plt.plot(lasso.coef_, 's', label="Lasso alpha=1")
# 绘制alpha值等于0.1时的模型系数
plt.plot(lasso01.coef_, '^', label="Lasso alpha=0.1")
# 绘制alpha值等于0.0001时的模型系数
plt.plot(lasso00001.coef_, 'v', label="Lasso alpha=0.0001")
# 绘制alpha值等于0.1时的岭回归模型系数作为对比
plt.plot(ridge01.coef_, 'o', label="Ridge alpha=0.1")
plt.legend(ncol=2, loc=(0,1.05))
# 限制 y 轴的范围
#plt.ylim(-25,25)
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
```

实际应用中，通常先决定使用 L1 正则化（特征很多，不是每个特征都对结果有影响）还是 L2 正则化（特征不多每个都重要）。

通常岭回归模型表现会更好。

scikit-learn 还提供了 **弹性网模型（ElasticNet）** ：综合了套索回归和岭回归的惩罚因子（需要调节 L1 正则化参数和 L2 正则化参数）。实践中是**更好**的选择。

```python
import numpy as np
# 将X, y赋值为np数组
X = np.array([[0, 1, 0, 1],
              [1, 1, 1, 0],
              [0, 1, 1, 0],
              [0, 0, 0, 1],
              [0, 1, 1, 0],
              [0, 1, 0, 1],
              [1, 0, 0, 1]])
y = np.array([0, 1, 1, 0, 1, 0, 0])
# 对不同分类计算每个特征为1的数量
counts = {}
for label in np.unique(y):
    # y == label 将数组 y 中的每个值与 label 对比生成一个新的布尔数组
    # X[y == label] 是使用数组的布尔索引，X 中 True 对应的元素被取出构成新的数组；即，筛选label对应特征数据
    # sum(axis=0) 按列对二维数组求和——放回一维数组
    counts[label] = X[y == label].sum(axis=0)
# 打印计数结果
print("feature counts:\n{}".format(counts))
```

```python
# 导入贝努利贝叶斯
from sklearn.naive_bayes import BernoulliNB

clf = BernoulliNB()
clf.fit(X,y)
# 预测数据
Next_Days = [[0,0,1,0],
           [1,1,0,1]]
# 分类预测结果
print(clf.predict(Next_Days))
# 预测分类的概率
print(clf.predict_proba(Next_Days))
```

**scikit-learn 提醒朴素贝叶斯虽然是相当好的分类器，但是预测概率数值的参考意义不大**
